### Some functions

```python
s = data['RunTime'].sum()
t = data['RunTime'].count()

avg = s / t
print(f"Average movies lenght: {avg}")
```

### Select and search

```python
good_movies = data[data['RATING'] >= 8.5]
good_movies
```

```python
max_len = data[data['RunTime']==data['RunTime'].max()]
max_len
```

```python
data[data['RATING'] > 9][data['RunTime'] > 60][data['YEAR'] == '2020']
```

#### Plotting

```python
data["RATING"].plot(kind='hist', bins=50, title='Movies rating', color='red', grid=True)
```
